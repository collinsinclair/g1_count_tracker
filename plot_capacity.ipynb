# Average Hourly Occupancy of G1 Climbing+Fitness by Day of Week

## Imports

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
%matplotlib inline
```

## Setup

```python
data = pd.read_csv('/Users/collinsinclair/Documents/Other/g1_capacity/g1_occupancy.csv')
timestamps = pd.to_datetime(data['timestamp'])
hours = timestamps.dt.hour
weekdays = timestamps.dt.dayofweek
counts = data['count']
    
mon = weekdays == 0
tue = weekdays == 1
wed = weekdays == 2
thu = weekdays == 3
fri = weekdays == 4
sat = weekdays == 5
sun = weekdays == 6

daily_hours = range(24)
open_hours = range(6,23,1)
```

## Overall Hourly Average

```python
hourly_count_means = []
for hour in open_hours:
    current_hour = hours == hour
    current_count = counts[current_hour]
    current_hour_mean_count = np.mean(current_count)
    hourly_count_means.append(current_hour_mean_count)
```

```python
# plt.xkcd()
plt.plot(open_hours, hourly_count_means)
plt.xticks(ticks = open_hours)
plt.xlabel('Hour of Day')
plt.ylabel('Average Occupancy')
plt.title('Overall Hourly Average Occupancy at G1 Climbing+Fitness')
plt.savefig('overall_hourly_avg.pdf')
```
